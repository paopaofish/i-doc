```python
# b
class Queue:

    # To initialize the object.
    def __init__(self):
        self.QueueArray = []  ##string
        self.HeadPointer = 0  ##integer
        self.TailPointer = 0  ##integer
        self.NumberItems = 0  ##integer
        self.capacity = 10

    # Function to insert an element
    # at the rear of the queue
    def Enqueue(self, DataToAdd):
        # Check queue is full or not
        if self.NumberItems >= self.capacity:
            print("\nQueue is full")
            print("Failed to add {}".format(DataToAdd))
            return False
        # Insert element at the rear
        self.QueueArray.insert(self.TailPointer, DataToAdd)
        print("{} added successful".format(DataToAdd))
        self.TailPointer += 1
        self.NumberItems = self.NumberItems + 1
```

```python
# c
class Queue:

    # To initialize the object.
    def __init__(self):

        self.QueueArray = []  ##string
        self.HeadPointer = 0  ##integer
        self.TailPointer = 0  ##integer
        self.NumberItems = 0  ##integer
        self.capacity = 10

    # Function to insert an element
    # at the rear of the queue
    def Enqueue(self, DataToAdd):

        # Check queue is full or not
        if self.NumberItems >= self.capacity:
            print("\nQueue is full")
            print("Failed to add {}".format(DataToAdd))
            return False
        # Insert element at the rear
        self.QueueArray.insert(self.TailPointer, DataToAdd)
        print("{} added successful".format(DataToAdd))
        self.TailPointer += 1
        self.NumberItems = self.NumberItems + 1

    # Print the data pointed to by HeadPointer
    def Dequeue(self):
        # If queue is empty
        if self.NumberItems == 0:
            print("Queue is empty")
            return False
        ReturnValue = self.QueueArray[self.HeadPointer]
        self.HeadPointer = self.HeadPointer + 1
        self.NumberItems = self.NumberItems - 1
        print("the queueDequeue output value is {}".format(ReturnValue))
        return ReturnValue
```

```python
# d 1
class Queue:

    # To initialize the object.
    def __init__(self, c):

        self.QueueArray = []  ##string
        self.HeadPointer = 0  ##integer
        self.TailPointer = 0  ##integer
        self.NumberItems = 0  ##integer
        self.capacity = c

    # Function to insert an element
    # at the rear of the queue
    def Enqueue(self, DataToAdd):

        # Check queue is full or not
        if self.NumberItems >= self.capacity:
            print("\nQueue is full")
            print("Failed to add {}".format(DataToAdd))
            return False
        # Insert element at the rear
        self.QueueArray.insert(self.TailPointer, DataToAdd)
        print("{} added successful".format(DataToAdd))
        self.TailPointer += 1
        self.NumberItems = self.NumberItems + 1

    # Print the data pointed to by HeadPointer
    def Dequeue(self):
        # If queue is empty
        if self.NumberItems == 0:
            print("Queue is empty")
            return False
        ReturnValue = self.QueueArray[self.HeadPointer]
        self.HeadPointer = self.HeadPointer + 1
        self.NumberItems = self.NumberItems - 1
        print("the queueDequeue output value is {}".format(ReturnValue))
        return ReturnValue

    # Function to print queue elements
    def queueDisplay(self):

        if self.NumberItems == 0:
            print("\nQueue is Empty")
        # print elements
        for i in self.QueueArray:
            print(i, end=' ')
```

```python
# d 2
# take as input 11 string values from the user
# capacity 11
q = Queue(11)
for i in ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J","K"]:
    q.Enqueue(i)
q.Dequeue()
q.Dequeue()
```
